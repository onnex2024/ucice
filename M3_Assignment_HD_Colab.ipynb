# Task
Load the datasets "/content/DIA_trainingset_RDKit_descriptors.csv" and "/content/DIA_testset_RDKit_descriptors.csv", train a decision tree model on the training data, and evaluate it on the test data.

## Load data

### Subtask:
Load the training and testing datasets into pandas DataFrames.


**Reasoning**:
The first step is to import the pandas library and load the training and testing datasets into pandas DataFrames as instructed.



```python
import pandas as pd

train_df = pd.read_csv("/content/DIA_trainingset_RDKit_descriptors.csv")
test_df = pd.read_csv("/content/DIA_testset_RDKit_descriptors.csv")

print("Training data head:")
display(train_df.head())

print("\nTesting data head:")
display(test_df.head())
```

## Prepare data

### Subtask:
Separate features and target variables for both training and testing datasets.


**Reasoning**:
Separate the features and target variables for both the training and testing datasets as instructed.



```python
X_train = train_df.drop(['Label', 'SMILES'], axis=1)
y_train = train_df['Label']

X_test = test_df.drop(['Label', 'SMILES'], axis=1)
y_test = test_df['Label']

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

## Train model

### Subtask:
Train a decision tree classifier model using the training data.


**Reasoning**:
Train a decision tree classifier model using the training data.



```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate a DecisionTreeClassifier object
dt_model = DecisionTreeClassifier(random_state=42)

# Train the decision tree model on the training data
dt_model.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```

## Evaluate model

### Subtask:
Evaluate the trained model using the testing data to assess its performance.


**Reasoning**:
Import the necessary evaluation metrics and make predictions on the test data.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = dt_model.predict(X_test)
```

**Reasoning**:
Calculate and print the evaluation metrics.



```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```
